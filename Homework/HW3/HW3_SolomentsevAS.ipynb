## Упражнение 1

1. Что такое доверительный интервал?

4. Пусть $X_1, X_2, X_3 ∼ N(0, 1)$. Какое распределение будет у $Z = \frac{ \sqrt{2}X_1 } {\sqrt{ X^2_2 + X_3^2} }$?
5. Какие плюсы и недостатки есть у бутстрэпа? Как думаете, почему этот метод так популярен?

### Ответ:
1. Доверительный интервал – это интервал, который с заданной вероятностью ($ 1 - \alpha $) покрывает оценки для выборки. 

$ \mathbb{P}(\hat{\theta}_-(X) < \theta < \hat{\theta}_+(X)) \geq 1 - \alpha $

2. Асимтотических доверительный интервал отличается от неасимптотического тем, что асимптотический доверительный интервал получается при использовании одной из предельных теорем, где $n \rightarrow \infty$ 

3. квантиль стандартного нормального распределения уровня уровня $0.1$ больше, потому что он ближе расположен к 0 (математическому ожиданию). 

4. У $Z$ будет распределение Бернулли с параметрами $(-1;1)$, так как знак будет зависеть от случайной величины из стандартного нормального распределения $X_1$ в числителе, а в знаменателе будет всегда положительное число $\sqrt{ X^2_2 + X_3^2}$ при этом скорее всего равное $\sqrt{2}X_1$

5. Плюсы бутстрэпа:
    * быстрота реализации и простота использования
    * позволяет использовать для сложных оценок
    * может быть более точной, если нет информации о точной функции распределения
    
Недостатки бутстрэпа:
   * требует много вычислительных ресурсов и времени 
   * может не давать теоретических гарантий

## Упражнение 2

В этом упражнении мы обсудим правила двух и трех сигм для произвольной случайной величины. Пусть $X$ имеет некоторое распределение с математическим ожиданием $\mathbb{E}X = a$ и конечной дисперсией $\mathrm{Var}(X) = σ^2 < ∞$. С помощью неравенства Чебышёва покажите, с какой вероятностью эта случайная величина лежит в отрезках $[a − 2σ, a + 2σ]$ и $[a − 3σ, a + 3σ]$. Сравните полученные вероятности с соответствующими вероятностями для стандартного нормального распределения $N (0, 1)$. Какой вывод можно сделать?

### Ответ:

Согласно неравенству Чебышёва вероятность того, что центрированная случайная величина $X$ с некоторым математическим ожиданием $\mu$ и конечной дисперсией $\sigma^2 < \infty$ по абсолютной величине превзойдет любое наперед заданное положительное число $k$, будет ограничено сверху отношением дисперсии к квадрату заданного значения.

$\mathbb{P}(|X - \mu| \geq k) \leq \frac{\sigma^2}{k^2}$

Подставим данные в условии задачи значения и получим:
1. Для случая $[a − 2σ, a + 2σ]$ вероятность составит $\mathbb{P}(|X - \mu| \geq 2\sigma) \leq \frac{\sigma^2}{(2\sigma)^2} \leq \frac{1}{4}$
2. Для случая $[a − 3σ, a + 3σ]$ вероятность составит $\mathbb{P}(|X - \mu| \geq 3\sigma) \leq \frac{\sigma^2}{(3\sigma)^2} \leq \frac{1}{9}$

Следовательно, вероятность того, что случайная величина лежит в отрезке $[a − 2σ, a + 2σ]$ составляет более $75$%, а в отрезке $[a − 3σ, a + 3σ]$ составляет более $88,88$%.

Для стандартного нормального распределения $N (0, 1)$ вероятность того, что случайная величина лежит в отрезке $[a − 2σ, a + 2σ]$ составляет $97,725$%, а в отрезке $[a − 3σ, a + 3σ]$ составляет $99,865$%.

**Вывод**: результат для стандартного нормального распределения входит в область, полученную с помощью неравенства Чебышева

## Упражнение 3

Пусть имеется реализация выборки $x_1, \dots , x_n$ из равномерного распределения на $[0, \theta]$ с неизвестным параметром $\theta$ > 0.

(1) Постройте точный доверительный интервал для параметра $\theta$ уровня доверия $1 − \alpha$ с помощью статистики $\hat{\theta}_1 = x_{(n)} = \max\{x_1,\dots,x_n\}$. Используйте тот факт, что распределение максимума из $n$ независимых равномерно распределённых на отрезке $[0, 1]$ случайных величин имеет функцию распределения

$ F(u) = \begin{cases} 0, & если \ u < 0 \\ u^n, & если \ u \in [0,1] \\ 1, & если \ u > 1 \end{cases} $

(2) Постройте асимптотически точный доверительный интервал для неизвестного параметра $\theta$ уровня доверия $1 − \alpha$, используя статистику $\hat{\theta}_1 = 2 \overline{x} = 2(x_1 + \dots + x_n)/n$.

### Ответ:

(1)
Найдем функцию $G(X,\theta)$, функция распределения которой не зависит от неизвестного параметра $\theta$.
Пусть $Y_{(n)} = \frac{X_{(n)}}{\theta} = \frac{ \max \{ x_1, \dots , x_n \} } {\theta} = \max \{ y_1, \dots , y_n \} = G(X, \theta)$

Найдем $g_1$ и $g_2$ – квантили распределения $G(X, \theta)$, для которых $\mathbb{P} (g_1 < G(X, \theta) < g_2)$

Подставим $\frac{X_{(n)}}{\theta}$ вместо $G(X,\theta)$ и разрешим относительно $\theta$. Отсюда получаем $\mathbb{P} (g_1 < \frac{X_{(n)}}{\theta} < g_2) = \mathbb{P} \left(\frac{X_{(n)}}{g_2} < \theta < \frac{X_{(n)}}{g_1}\right)$

Также найдем значения $g_1$ и $g_2$ исходя из определения точного доверительного интервала. Тогда получим $\mathbb{P} (g_1 < G(X, \theta) < g_2) = F(g_2) - F(g_1) = 1 - \alpha $. Отсюда удобно выбрать такие $g_1$ и $g_2$, чтобы $F(g_2) = 1$ и $F(g_1) = \alpha$. При заданной функции распределения получается $g_2 = 1$, а $g_1 = \sqrt[n]{\alpha}$ 

Подставим полученные значения, получаем $1 - \alpha = \mathbb{P} \left(\frac{X_{(n)}}{1} < \theta < \frac{X_{(n)}}{\sqrt[n]{\alpha}}\right)$. Следовательно точный доверительный инвервал $ \left( X_{(n)} , \frac{X_{(n)}}{\sqrt[n]{\alpha}} \right) $



(2)
Воспользуемся центральной предельной теоремой, чтобы найти функцию $G(X,\theta)$,  которая бы сходилась к известной случайной величине $Z$, не зависящей от неизвестного параметра $\theta$.

$ \frac{ n\overline{X} - \mathbb{E}(n\overline{X}) }{ \sqrt{\mathrm{Var}(n\overline{X}) } } = \sqrt{n} \frac{ \overline{X} - \theta / 2 }{ \theta / \sqrt{12}} \approx N(0,1) $

По центральной предельной теореме распределение случайной величины слабо сходится к стандартному нормальному закону, а $2\overline{X}$ сходится по вероятности к $\theta$. Следовательно, можем записать, что для произвольных $a < b$

$\mathbb{P} \left( a <  \sqrt{n} \frac{ \overline{X} - \theta }{ 2 \overline{X} / \sqrt{12}}  < b  \right) \rightarrow \mathbb{P} \left( a < Z < b  \right)$ при $n \rightarrow \infty$

Возьмем границы интервала  $a = c_{α/2} = -c_{1−α/2} , b = c_{1−α/2}$, тогда
$\mathbb{P} \left( -c_{1-α/2} <  \sqrt{n} \frac{ \overline{X} - \theta/2}{ \overline{X} / \sqrt{3}}  < c_{1−α/2}  \right) \rightarrow 1 - \alpha$ при $n \rightarrow \infty$

Разрешив относительно $\theta$, получим
$\mathbb{P} \left(2\overline{X} - \frac{c_{1-α/2}}{ \sqrt{3n} } <  \theta  < 2\overline{X} + \frac{c_{1-α/2}}{ \sqrt{3n} }  \right) \rightarrow 1 - \alpha$

Таким образом асимптотический доверительный интервал составляет $\left(2\overline{X} - \frac{c_{1-α/2}}{ \sqrt{3n} } , 2\overline{X} + \frac{c_{1-α/2}}{ \sqrt{3n} }  \right)$

## Упражнение 4

Затем мы генерируем новую выборку из равномерного распределения на $[0, 2\overline{x}]$ и оцениваем с ее помощью какую-то величину, которая нам интересна (то есть мы используем параметрический бутстрэп для оценки этой величины). Какую дисперсию будут иметь случайные величины из равномерного распределения на $[0, 2\overline{x}]$? Сравните ее с дисперсией равномерного распределения на $[0, \theta]$. Какой вывод можно сделать?

### Ответ:

Дисперсия случайной величины для равномерного распределения равна $\frac {(b-a)^2}{12}$, где $a$ – начало отрезка, а $b$ – конец отрезка.
Следовательно дисперсия для $[0, \theta]$ будет $\frac {\theta^2}{12}$, а для $[0, 2\overline{x}]$ будет $\frac {(2\overline{x})^2}{12} = \frac {\overline{x}^2}{3}$.

## Упражнение 5

Рассмотрим нормальное распределение $N (θ, σ^2)$ с неизвестными параметрами $θ ∈ R$ и $σ^2 > 0$. В этой задаче мы численно сравним следующие доверительные интервалы для параметра $θ$: 1) теоретический, 2) на основе параметрического бутстрэпа, 3) на основе непараметрического бутстрэпа. Для этого:

(1) сгенерируйте неизвестный параметр θ из равномерного распределения на $[10,20]$ и $σ$ — из равномерного распределения на $[1, 2]$;

(2) сгенерируйте выборку из нормального распределения $N (θ, σ^2)$ размера $n = 50$;

(3) постройте доверительные интервалы 1), 2), 3) уровня доверия $1 − α$;

(4) сравните длины полученных доверительных интервалов;

(5) проведите этот эксперимент на $10 000$ выборках; с какой частотой $θ$ попадает в полученные доверительные интервалы?

```python
import numpy as np
import scipy.stats as stats
from math import sqrt
```

```python
# Параметры из задачи
n = 50
alpha = 0.05 # произвольно заданный уровень 
experiment = 10000

# Сгенерируем неизвестный параметр тета и сигма
np.random.seed(321)

theta = np.random.uniform(10, 20)
sigma = np.random.uniform(1, 2)
print(f'Истинное значение тета составляет {theta}, а сигма – {sigma}.\n')

# Сгенерируем выборку из нормального распределения размера 
sample = np.random.normal(loc= theta, scale= sigma, size=n)

# Теоретический доверительный интервал
CI_Theoretical = [sample.mean() - stats.norm.ppf(1-alpha/2) * sample.std(ddof=1)/sqrt(n), 
                  sample.mean() + stats.norm.ppf(1-alpha/2) * sample.std(ddof=1)/sqrt(n)]
print(f'Теоретический доверительный интервал составляет: {CI_Theoretical}.\n')

# Доверительный интервал на основе параметрического бутстрепа
pb_est = np.apply_along_axis(np.mean, 1, np.random.normal(sample.mean(), sample.std(ddof=1), size=[n,n]))
CI_Bootstrap_Parametric = [np.quantile(pb_est, alpha/2), np.quantile(pb_est, 1-alpha/2)]
print(f'Доверительный интервал на основе парметрического бустрэпа равен {CI_Bootstrap_Parametric}.\n')

# Доверительный интервал на основе непараметрического бутстрепа
npb_est = np.apply_along_axis(np.mean, 1, np.random.choice(sample, size=[n,n]))
CI_Bootstrap_Nonparametric = [np.quantile(npb_est, alpha/2), np.quantile(npb_est, 1-alpha/2)]
print(f'Доверительный интервал на основе непарметрического бустрэпа равен {CI_Bootstrap_Nonparametric}.\n')

# Сравнение длин доверительных интервалов
CITL = CI_Theoretical[1]-CI_Theoretical[0]
CIPBL = CI_Bootstrap_Parametric[1]-CI_Bootstrap_Parametric[0]
CINPBL = CI_Bootstrap_Nonparametric[1]-CI_Bootstrap_Nonparametric[0]
res = list(zip(['теоретического интервала', 'парметрического бустрэпа', 'непарметрического бустрэпа'], 
               [CITL, CIPBL, CINPBL]))
res.sort(key=lambda x: x[1], reverse=True)
print(f'Длина теоретического доверительного интервала: {CITL}.')
print(f'Длина доверительного интервала на основе парметрического бустрэпа: {CIPBL}.')
print(f'Длина доверительного интервала на основе непарметрического бустрэпа: {CINPBL}.')
print(f'Наиболее широкий доверительный интервал у {res[0][0]}. Наиболее узкий у {res[-1][0]}.\n')


# Эксперимент на 10 тыс. выборок
theor_frac, pb_frac, npb_frac = np.zeros(experiment), np.zeros(experiment), np.zeros(experiment)

def ci_theor(sample, alpha, n):
    CI_Theoretical = [sample.mean() - stats.norm.ppf(1-alpha/2) * sample.std(ddof=1)/sqrt(n), 
                      sample.mean() + stats.norm.ppf(1-alpha/2) * sample.std(ddof=1)/sqrt(n)]
    return CI_Theoretical

def ci_pb(sample, alpha, n):
    pb_est = np.apply_along_axis(np.mean, 1, np.random.normal(sample.mean(), sample.std(ddof=1), size=[n,n]))
    CI_Bootstrap_Parametric = [np.quantile(pb_est, alpha/2), np.quantile(pb_est, 1-alpha/2)]
    return CI_Bootstrap_Parametric

def ci_npb(sample, alpha, n):
    npb_est = np.apply_along_axis(np.mean, 1, np.random.choice(sample, size=[n,n]))
    CI_Bootstrap_Nonparametric = [np.quantile(npb_est, alpha/2), np.quantile(npb_est, 1-alpha/2)]
    return CI_Bootstrap_Nonparametric

for i in range(experiment):
    i_sample = np.random.normal(loc=theta, scale=sigma, size=n)
    
    CI_Theoretical = ci_theor(i_sample, alpha, n)
    theor_frac[i] = ((theta >= CI_Theoretical[0]) and (theta <= CI_Theoretical[1]))
    
    CI_Bootstrap_Parametric = ci_pb(i_sample, alpha, n)
    pb_frac[i] = ((theta >= CI_Bootstrap_Parametric[0]) and (theta <= CI_Bootstrap_Parametric[1]))
    
    CI_Bootstrap_Nonparametric = ci_npb(i_sample, alpha, n)
    npb_frac[i] = ((theta >= CI_Bootstrap_Nonparametric[0]) and (theta <= CI_Bootstrap_Nonparametric[1]))

print(f'''Частота попадания истинного параметра в доверительный интервал:
- для теоретического доверительного интервала {np.mean(theor_frac)};
- для параметрического бутстрэп доверительного интервала {np.mean(pb_frac)};
- для непараметрического бутстрэп доверительного интервала {np.mean(npb_frac)}.''')
```

В файле `Banner_small.npy` даны две выборки из распределения Бернулли. Они соответствуют кликам по двум рекламным баннерам в Интернете. Постройте доверительные интервалы для частоты кликов (то есть параметра «успеха»). Какой из изученных вариантов доверительных интервалов лучше применять в этой задаче? Как думаете, можно ли сказать, что один из баннеров лучше, чем другой? Если да, то с какой вероятностью? Проверьте свой вывод на выборке большего размера, которая записана в файл `Banner_large.npy`.

```python
import statsmodels.stats.proportion

sample1, sample2 = np.load('HW_conf_int_data/Banner_small.npy')

alpha = 0.05 # произвольно заданный уровень 
step = 0.01 # произвольный шаг

# Доверительные интервалы для каждой из двух выборок

CI_Normal_1 = statsmodels.stats.proportion.proportion_confint(sum(sample1), len(sample1), 0.05, method = 'normal')
CI_Normal_2 = statsmodels.stats.proportion.proportion_confint(sum(sample2), len(sample2), 0.05, method = 'normal')

CI_Wilson_1 = statsmodels.stats.proportion.proportion_confint(sum(sample1), len(sample1), 0.05, method = 'wilson')
CI_Wilson_2 = statsmodels.stats.proportion.proportion_confint(sum(sample2), len(sample2), 0.05, method = 'wilson')

print(f'''Асимптотический доверительный интервал на основе ЦПТ для первой выборки равен {CI_Normal_1}.
Асимптотический доверительный интервал на основе ЦПТ для второй выборки равен {CI_Normal_1}.

Доверительный интервал Уилсона для первой выборки равен {CI_Wilson_1}.
Доверительный интервал Уилсона равен для второй выборки {CI_Wilson_2}.

Доверительный интервал Уилсона является улучшением по сравнению с асимптотическим доверительным интервалом 
на основе ЦПТ, потому что он не страдает от проблем с превышением и интервалами нулевой ширины, которые влияют 
на нормальный интервал по ЦПТ, и его можно безопасно использовать с небольшими выборками и искаженными 
наблюдениями. Например, как в случае представленной задачи.\n''')


# Доверительный интервал для разности долей в случае двух выборок

def difference_in_proportions(sample1, sample2, alpha = 0.05):    
    z = stats.norm.ppf(1 - alpha / 2.)   
    p1 = sample1.mean()
    p2 = sample2.mean()
    
    l = (p1 - p2) - z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    r = (p1 - p2) + z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    
    return (l, r)

left, right = difference_in_proportions(sample1, sample2, alpha)
if ((left < 0) & (right < 0)) or ((left > 0) & (right > 0)):
    print(f'''Так как интервал не пересекает ноль, значит разность статистически значима. Следовательно с
вероятностью {100*(1-alpha)}% можно утверждать, что один из баннеров лучше, чем другой.\n''')
else:
    print(f'''Так как интервал пересекает ноль, значит разность статистически не значима. Следовательно нельзя с
вероятностью {100*(1-alpha)}% утверждать, что один из баннеров лучше, чем другой.\n''')
    
    
# Вероятность с которой можно утверждать, что один из баннеров лучше

alpha_test = alpha
while alpha_test < 1:
    l,r = difference_in_proportions(sample1, sample2, alpha_test)
    if ((l < 0) & (r < 0)) or ((l > 0) & (r > 0)):
        print(f'''Так как интервал {l,r} не пересекает ноль, значит разность статистически 
значима. Следовательно с вероятностью {100*(1-alpha_test)}% можно утверждать, что один из баннеров лучше, чем другой.\n''')
        break
    alpha_test += step
    
    
# Проверка на большом объеме выборок

sample1_re, sample2_re = np.load('HW_conf_int_data/Banner_large.npy')
left_re, right_re = difference_in_proportions(sample1_re, sample2_re, alpha)

print(f'''Проверим выводы на выборках большего объема:''')

if ((left_re < 0) & (right_re < 0)) or ((left_re > 0) & (right_re > 0)):
    print(f'''Так как интервал не пересекает ноль, значит разность статистически значима. Следовательно с
вероятностью {100*(1-alpha)}% можно утверждать, что один из баннеров лучше, чем другой.\n''')
else:
    print(f'''Так как интервал пересекает ноль, значит разность статистически не значима. Следовательно нельзя с
вероятностью {100*(1-alpha)}% утверждать, что один из баннеров лучше, чем другой.\n''')
```
