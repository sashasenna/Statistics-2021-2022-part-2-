## 1 Про корреляции

(10 баллов). Ответьте на следующие вопросы:
    
1) Что можно сказать про случайные величины $X$ и $Y$ , если $Corr(X, Y ) = 1$? 

2) Чем отличаются коэффициенты корреляции Пирсона и Спирмена? В каких случаях лучше пользоваться коэффициентом корреляции Спирмена?

3) Дана случайная величина $X$, равномерно распределенная на отрезке $[−1,1]$ и величина $Y = |X|$. Чему равен их коэффициент корреляции (вычислите $Cov(X, Y )\big/ \sqrt{ \left( Var(X) Var(Y) \right) }$ ?

4) Дана случайная величина $X$, равномерно распределенная на отрезке $[−1,1]$ и величина $Y = X^2$. Сгенерируйте выборку длины 1000 величины X и на основании нее получите соответствующую выборку для величины $Y$. По полученным выборка посчитайте коэффициенты корреляции Пирсона и Спирмена.

### Ответ:

1. Между случайными величинами $X$ и $Y$ существует точная линейная связь, такая что $ Y = aX + b $, для некоторых $a > 0$ и $b \in \mathbb{R}$ 

2. Отличие состоит в методе расчета, в коэффициенте корреляции Пирсона используются значения случайных величин $X$ и $Y$ (параментрический метод), а в коэффициенте корреляции Спирмена – упорядоченные ранги этих случайных величин (непараметрический метод). Целесообразнее использовать коэффициент корреляции Спирмена в случае наличия выбросов и может быть использован с качественными показателями, если такие показатели можно ранжировать.

3. Если случайная величина $X$ равномерно распределена на отрезке $[−1,1]$, то величина $Y = |X|$ равномерно распределена на отрезке $[0,1]$.
Вычислим $Cov(X, Y)$: 

$Cov(X, Y) = \mathbb{E}[XY] - \mathbb{E}[X] \mathbb{E}[Y]$

$\mathbb{E}[X] = \frac{-1 + 1} {2} = 0$

$\mathbb{E}[Y] = \frac{0 + 1} {2} = 0.5 $

Произведение случайных величин $X$ и $Y$ равномерно распределено на отрезке $[−1,1]$. Следовательно $\mathbb{E}[XY] = \frac{-1 + 1} {2} = 0$.

Тогда $Cov(X, Y) = \mathbb{E}[XY] - \mathbb{E}[X] \mathbb{E}[Y] = 0 - 0 \cdot 0.5 = 0$

Вычислим $Var(X)$ и $Var(Y)$:

$Var(X) = \frac{[1 - (-1)]^2} {12} = \frac{4}{12} = \frac{1}{3}$

$Var(Y) = \frac{[1 - 0]^2} {12} = \frac{1}{12}$

Отсюда $Corr(X, Y ) =  \frac{0}{\sqrt{\frac{1}{3} \cdot \frac{1}{12}}} = 0$

```python
import numpy as np
from scipy import stats

# 4

np.random.seed(132)
X = np.random.uniform(-1,1, size=1000)
Y = X ** 2

Pearson_corr = stats.pearsonr(X, Y)
Spearman_corr = stats.spearmanr(X, Y)

print(f'Коэффициент корреляции Пирсона равен {Pearson_corr[0]}, a p-value – {Pearson_corr[1]}.')
print(f'Коэффициент корреляции Спирмена равен {Spearman_corr[0]}, a p-value – {Spearman_corr[1]}.')
```

**Упражнение 2** (20 баллов). Сгенерируйте 100 выборок длины 1000 из распределения $N(0,1)+exponential(1)$. Для каждой из выборок $X_1,X_2 \dots X_{100}$ постройте по 5 выборок $Y$ вида:

$Y^{(1)} =X_i^2 + 0.1 N(0,1)$

$Y^{(2)} =\sqrt{|X_i|} + 0.1 N(0,1)$

$Y^{(3)} =X_i * \sin(X_i) + 0.1 N(0,1) $

$Y^{(4)} =X_i^3 + 0.1 N(0,1)$

$Y^{(5)} =X_i^3 \cos(X_i) + 0.1 N(0,1)$

для каждой пары выборок $X_i$, соответствующая ей $Y_i$, посчитать коэффициенты корреляции Спирмена и Кендалла. (Всего должно получиться 500 значений для Спирмена и 500 значений для Кендалла). Нарисуйте диаграмму рассеяния (scatterplot), где по оси $X$ будет корреляция Спирмена, а по оси $Y$ - Кендалла. Можно ли сказать на основании собранных нами данных, что эти коэффициенты корреляции как-то связаны друг с другом?

```python
# Генерируем 100 выборок длины 1000 из распределения N(0,1) + Exp(1)

N_samples = 100
sample_size = 1000

X = stats.norm(loc= 0, scale=1).rvs(size=[N_samples, sample_size]) + stats.expon(loc= 0, scale= 1).rvs(size=[N_samples, sample_size])
```

```python
# Генерируем 5 новых масивов по 100 выборок длины 1000 

Y_1 = X ** 2 + 0.1 * stats.norm(loc= 0, scale= 1).rvs(size=[N_samples, sample_size])
Y_2 = np.sqrt(np.abs(X)) + 0.1 * stats.norm(loc= 0, scale= 1).rvs(size= [N_samples, sample_size])
Y_3 = X * np.sin(X) + 0.1 * stats.norm(loc= 0, scale= 1).rvs(size= [N_samples, sample_size])
Y_4 = X ** 3 + 0.1 * stats.norm(loc= 0, scale= 1).rvs(size= [N_samples, sample_size])
Y_5 = X ** 3 * np.cos(X) + 0.1 * stats.norm(loc= 0, scale= 1).rvs(size= [N_samples, sample_size])


# Определяем коэффициенты корреляции для каждой пары из X и Y по 100 выборок длины 1000 
kendall_tau = []
spearmen_rho = []
kendall_p = []
spearmen_p = []


for Y in [Y_1, Y_2, Y_3, Y_4, Y_5]:
    for i in range(X.shape[0]):
        tau, p = stats.kendalltau(X[i], Y[i])
        
        kendall_tau.append(tau)
        kendall_p.append(p)
        
        rho, p_val = stats.spearmanr(X[i], Y[i])
        spearmen_rho.append(rho)
        spearmen_p.append(p_val)
```

```python
# Рисуем графики

fig, ax = plt.subplots(figsize=(12,12), ncols=2, nrows=3, constrained_layout=True)

ax[0][0].set_title('All sameples')
ax[0][0].set_xlabel('Spearman correlation coefficient')
ax[0][0].set_ylabel('Spearman correlation coefficient')
ax[0][0].scatter(x= spearmen_rho, y= kendall_tau)

ax[0][1].set_title('X and $Y_1$ samples')
ax[0][1].set_xlabel('Spearman correlation coefficient')
ax[0][1].set_ylabel('Spearman correlation coefficient')
ax[0][1].scatter(x= spearmen_rho[:100], y= kendall_tau[:100])

ax[1][0].set_title('$X$ and $Y_2$ samples')
ax[1][0].set_xlabel('Spearman correlation coefficient')
ax[1][0].set_ylabel('Spearman correlation coefficient')
ax[1][0].scatter(x= spearmen_rho[100:200], y= kendall_tau[100:200])

ax[1][1].set_title('$X$ and $Y_3$ samples')
ax[1][1].set_xlabel('Spearman correlation coefficient')
ax[1][1].set_ylabel('Spearman correlation coefficient')
ax[1][1].scatter(x= spearmen_rho[200:300], y= kendall_tau[200:300])

ax[2][0].set_title('$X$ and $Y_4$ samples')
ax[2][0].set_xlabel('Spearman correlation coefficient')
ax[2][0].set_ylabel('Spearman correlation coefficient')
ax[2][0].scatter(x= spearmen_rho[300:400], y= kendall_tau[300:400])

ax[2][1].set_title('$X$ and $Y_5$ samples')
ax[2][1].set_xlabel('Spearman correlation coefficient')
ax[2][1].set_ylabel('Spearman correlation coefficient')
ax[2][1].scatter(x= spearmen_rho[400:], y= kendall_tau[400:])

plt.show()
```

**Ответ:** На основании собранных данных можно сказать, что эти коэффициенты корреляции Спирмана и Кенделла связаны друг с другом линейно и сонаправлены друг другу.

(80 баллов) (Обязательно попробовать) Представьте, что ваша компания занимается продажей товаров. Покупка устроена так: человек заходит на сайт (аналогично нашему показателю views), затем совершает покупку с некоторой вероятностью (аналогично нашему success-rate с бета-распределением) и далее совершает покупку (аналогично нашему clicks). Каждая покупка характеризуется каким-то "чеком" (стоимостью товара), его можно моделировать с помощью экспоненциального распределения вида $C e^{−λx}$ на интервале $[ 100, 2500 ]$ (подберите для этого распределения правильную нормировочную константу $C$ и некоторую разумную $λ$, помните - это распределение чеков на покупку каких-то недорогих, частых товаров).

Допустим, что в обеих группах базовый $\text{success-rate} = 0.03$ , а в группе B его $\text{uplift}$ равен $0.1$. Кроме того, в группе $\text{B}$ распределение чеков имеет меньшее значение $λ: λ_A = 1.2λ_B$.

Мы хотим измерить, значимо ли меняется значение 80 квантиля для чека пользователей в группе. Проверьте, какой из тестов для этого лучше подойдет: тест Манна-Уитни поверх бакетного преобразования или Пуассоновский бутстреп? Как их работоспособность зависит от скошенности распределений истинного CTR (бета-распределения) пользователей и показов (логнормального распределения). Для оценки можно использовать те же параметры, что и в лекции.

```python
experiment = 1000
sample_size = 5000

views = None # Lognormal
CTR = 0.03 # Beta() = success-rate
clicks = None # Binomial
price = None # Exponential -> [100, 2500]
lambda_B = 0.9 # Exponential parameter -> [100, 2500]
lambda_A = 1.2 * lambda_B
C = 100 # constant
uplift = 0.1
```

```python
# Генерируем распределения просмотров

mu = 1 # E[x] for Lognormal distribution
sigma = 1 # std for Lognormal distribution

views_A = (stats.lognorm(loc= 1, scale= mu, s= sigma).rvs(size=[experiment, sample_size])).astype(int)
views_A.shape, views_A.min(), views_A.max()
```

```python
import seaborn as sns

fig = plt.figure()
ax_views = fig.add_subplot()
views_distr = sns.histplot(views_A.ravel(),  # Теперь это массив длины experiments * N
                           bins=range(0, 30),
                           kde=False,
                           ax=ax_views,
                           stat='probability')
ax_views.set_xlim((0, 30))
views_99_percentile = np.percentile(views_A.ravel(), 99)
ax_views.set_title(f'Views, 99%-ile = {views_99_percentile:<7.1f}')
```

```python
# Генерируем распределение ctr

b = 1000 # beta for Beta()
a = CTR * b / (1 - CTR) # alpha for Beta()

CTR_A = stats.beta(a= a, b= b).rvs(size=[experiment, sample_size])
CTR_A.shape, CTR_A.max(), CTR_A.min(), CTR_A.mean(), np.median(CTR_A)
```

```python
fig = plt.figure()
ax_ctr = fig.add_subplot()
sns.histplot(CTR_A.ravel(),
             bins=np.linspace(0, 0.2, 100),
             ax=ax_ctr,
             kde=False,
             stat='probability')
ax_ctr.set_xlim((0, 0.1))
success_rate_std = CTR_A[:10].flatten().std()
ax_ctr.set_title(f'Ground truth user CTR, std = {success_rate_std:2.3f}')
```

```python
# Генерируем распределение кликов

clicks_A = stats.binom.rvs(views_A, CTR_A)
clicks_A.shape, clicks_A.max(), clicks_A.min(), (clicks_A == 0).sum() / (experiment * sample_size)
```

```python
(experiment * sample_size) * CTR
```

```python
# Генерируем распределение цен

prices_A = (C * stats.expon(loc= 1, scale= 1 / lambda_A).rvs(size=[experiment, sample_size])).round(decimals= 2)
prices_A.shape, prices_A.min(), (prices_A > 0).sum()
```

```python
# Считаем чеки
revenue_user_A = clicks_A * prices_A

revenue_user_A.shape
```
